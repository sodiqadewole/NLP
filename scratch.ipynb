```python
from sklearn.feature_extraction.text import CountVectorizer
import seaborn as sns

corpus = [
    'Time flies like an arrow', 
    'Fruit flies like a banana'
]

vocab = list(set(word for sentence in [sentence.split(" ") for sentence in corpus] for word in sentence))

one_hot_vectorizer = CountVectorizer(binary=True)
one_hot = one_hot_vectorizer.fit_transform(corpus).toarray()
sns.heatmap(one_hot, annot=True, cbar=False, xticklabels=vocab,
            yticklabels=['Sentence 1', 'Sentence 2'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import seaborn as sns
import numpy as np

corpus = [
    'Time flies flies like an arrow.', 
    'Fruit flies like a banana.'
]

tfidf_vectorizer = TfidfVectorizer()
tfidf = tfidf_vectorizer.fit_transform(corpus).toarray()
sns.heatmap(tfidf, annot=True, cbar=False, xticklabels=vocab,
            yticklabels=['Sentence 1', 'Sentence 2'])
```

```python
vec = tfidf_vectorizer.transform(["Fruit flies like an arrow"]).toarray().squeeze()
print(vec)
print()
vec = tfidf_vectorizer.transform(["Try a new sentence"]).toarray()
print(vec)
```

#### Tokenization

```python
import spacy

print(spacy.__version__)
nlp = spacy.load('en_core_web_sm')
sentence = "Mary, don't slap the green witch"
print([str(token) for token in nlp(sentence.lower())])
```

```python
from spacy.lang.yo import Yoruba

nlp = Yoruba()
nlp = spacy.blank("yo")
[str(token) for token in nlp("kini itumo owe yi?")]
```

```python
from nltk.tokenize import TweetTokenizer

tweet = u"Snow white and the Seven Degrees #MakeAMovieCold@midnight:-)"
tokenizer = TweetTokenizer()
print(tokenizer.tokenize(tweet.lower()))
```
